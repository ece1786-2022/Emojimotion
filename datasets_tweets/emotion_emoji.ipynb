```python
import pandas as pd
```

```python
url='https://raw.githubusercontent.com/abushoeb/EmoTag/master/data/EmoTag1200-scores.csv'
df = pd.read_csv(url, index_col=0)
emoji= pd.read_csv('emoji.csv')
df.head()
```

```python
emoji.head()
```

```python
from typing_extensions import TypeVarTuple
df1=pd.DataFrame(columns=['emoji','emoji_num','anger','anticipation','disgust','fear',
              'joy','sadness','surprise','trust' ])
for i in range(60):
  current=emoji['symbol'].loc[i]
  # print(i)
  # print(current)
  if current in df['emoji'].values:
    temp=df[df['emoji']==current]
    new_row = {'emoji':current,'emoji_num':i,'anger':float(temp['anger'].values),'anticipation':float(temp['anticipation']),'disgust':float(temp['disgust']),
               'fear':float(temp['fear']),'joy':float(temp['joy']),'sadness':float(temp['sadness']),
               'surprise':float(temp['surprise']),'trust':float(temp['trust'])}
    df1=df1.append(new_row,ignore_index=True)
  else:
    new_row = {'emoji':current,'emoji_num':i,'anger':0,'anticipation':0,'disgust':0,'fear':0,
              'joy':0,'sadness':0,'surprise':0,'trust':0}
    df1=df1.append(new_row,ignore_index=True)

    
```
